```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme()
sns.set(rc={'figure.figsize':(16,9)}, font_scale=2)
participants = [52,34,32,29,63,40,46,54,36,36,24,19,45,20,28,29,38,33,49,37]
df = pd.DataFrame({'Participants': participants})
ax = sns.histplot(data=df, x="Participants",bins=(19, 29, 39, 49, 59, 69))
ax.set_xticks([19, 29, 39, 49, 59, 69])
ax.set_ylabel('Frequency')
ax.set_xlabel('Age')
plt.savefig('survey_hist.png')
```
