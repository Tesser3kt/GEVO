```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme()
data = {
    'labels': ['Chickadee', 'Dove', 'Finch', 'Grackle', 'Sparrow', 'Starling'],
    'values': [.19, .06, .25, .13, .25, .13]
}
df = pd.DataFrame(data)
plt.figure(figsize=(6, 6))
plt.pie(df['values'], labels=df['labels'])
plt.title('Gardener Example')
plt.show()
```
